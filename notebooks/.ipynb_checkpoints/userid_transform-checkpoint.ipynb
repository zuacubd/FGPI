```python
import os
import sys

def get_content(path):
    with open(path, 'r') as fr:
        lines = fr.readlines()
    return lines
        
data_dir = "/projets/sig/mullah/nlp/fgpi/"
#retweetuserid_originaluserid = os.path.join(data_dir, 'data/processed/20000_UserRetweet_UserOriginal.txt')
#newretweetuserid_neworiginaluserid_path = os.path.join(data_dir, 'data/processed/20000_UserRetweetID_UserOriginalID_NewID.txt')
#newuserid_olduserid_path = os.path.join(data_dir, 'data/processed/20000_NewUserID_OldUserID.txt')
#olduserid_newuserid_path = os.path.join(data_dir, 'data/processed/20000_OldUserID_NewUserID.txt')

retweetuserid_tweetuserid_tweetid = os.path.join(data_dir, 'data/processed/20000_UserRetweet_UserOriginal_idOriginalTweet.txt')

newretweetuserid_neworiginaluserid_tweetid_path = os.path.join(data_dir, 'data/processed/20000_RetweetUserNewID_TweetUserNewID_TweetID.txt')
newuserid_olduserid_path = os.path.join(data_dir, 'data/processed/20000_NewUserID_OldUserID.txt')
olduserid_newuserid_path = os.path.join(data_dir, 'data/processed/20000_OldUserID_NewUserID.txt')

contents = get_content(retweetuserid_tweetuserid_tweetid)
print (len(contents))

ids = {}
for line in contents:
    parts = line.rstrip().split(',')
    #print (parts[0])
    #print (parts[1])
    if parts[0] in ids:
        ids[parts[0]] = ids[parts[0]] + 1
    else:
        ids[parts[0]] = 1
    
    if parts[1] in ids:
        ids[parts[1]] = ids[parts[1]] + 1
    else:
        ids[parts[1]] = 1

sorted_ids = sorted(ids.items(), key=lambda x: x[1], reverse=True)
print (len(sorted_ids))
#print (sorted_ids)
newuid_olduid = {}
olduid_newuid = {}

for newid in range(0, len(sorted_ids)):
    uid_degree = sorted_ids[newid]
    #print (newid)
    #print (uid_degree[0])
    #print (uid_degree[1])
    
    newuid_olduid[newid] = uid_degree[0]
    olduid_newuid [uid_degree[0]] = newid

print (len(newuid_olduid))
print (len(olduid_newuid))

new_ruid_ouid_list = []
contents = get_content(retweetuserid_tweetuserid_tweetid)
for line in contents:
    ruid = line.rstrip().split(',')[0]
    tuid = line.rstrip().split(',')[1]
    tid = line.rstrip().split(',')[2]
    
    new_ruid = olduid_newuid.get(ruid)
    new_ouid = olduid_newuid.get(tuid)
    
    new_ruid_ouid_list.append((new_ruid, new_ouid, tid))
    #print (str(new_ruid) + ' ->' + str(new_ouid))
print (len(new_ruid_ouid_list))

with open(newretweetuserid_neworiginaluserid_tweetid_path, 'w') as fw:
    for new_ruid_ouid in new_ruid_ouid_list:
        new_ruid = new_ruid_ouid[0]
        new_ouid = new_ruid_ouid[1]
        tid = new_ruid_ouid[2]
        fw.write(str(new_ruid) + ',' + str(new_ouid) + ',' + str(tid) + '\n')
    fw.close()

with open(newuserid_olduserid_path, 'w') as fw:
    for index, newuid in enumerate(newuid_olduid):
        olduid = newuid_olduid.get(newuid)
        fw.write(str(newuid) + ' ' + str(olduid) + '\n')
    fw.close()

with open(olduserid_newuserid_path, 'w') as fw:
    for index, olduid in enumerate(olduid_newuid):
        newuid = olduid_newuid.get(olduid)
        fw.write(str(olduid) + ' ' + str(newuid) + '\n')
    fw.close()
```
